## Criando Gráficos com Seaborn

```python
# carregando um dos datasets que vem com o Seaborn

dados = sns.load_dataset("tips")
dados.head()
```

```python
# método joinplot cria plot de 2 variáveis com gráficos bivariados e univariados

sns.jointplot(data = dados, x = "total_bill", y = "tip", kind = 'reg');
```

```python
# O método lmplot() cria plot com dados e modelos de regressão

sns.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")
```

```python
# gerando um dataframe com Pandas

df = pd.DataFrame()

df['idade'] = random.sample(range(20, 100), 30)
df['peso'] = random.sample(range(55, 150), 30)

df.head()
```

```python
# lmplot (linear model plot)

sns.lmplot(data = df, x = "idade", y = "peso") # fit_reg = True traça a linha de regressão e área da margem de erro
```

```python
# kdeplot

sns.kdeplot(df.idade)
```

```python
# kdeplot

sns.kdeplot(df.peso)
```

```python
# distplot

sns.distplot(df.peso)
```

```python
# Histograma

plt.hist(df.idade, alpha = .3) # histograma
sns.rugplot(df.idade) # rugas
```

```python
# Box Plot

sns.boxplot(df.idade, color = 'm')
```

```python
# Box Plot

sns.boxplot(df.peso, color = 'y')
```

```python
# Violin Plot

sns.violinplot(df.idade, color = 'g')
```

```python
# Violin Plot

sns.violinplot(df.peso, color = 'cyan')
```

```python
# Clustermap

sns.clustermap(df)
```

## Usando Matplotlib, Seaborn, NumPy e Pandas na Criação de Gráfico Estatístico

```python
# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

# Cria os dados para a variável flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})

dados.head()
```

```python
# Style
sns.set(style = "ticks")

# lmplot
sns.lmplot(x = 'altura', 
           y = 'peso', 
           data = dados, 
           hue = 'flag_fumante', 
           palette = ['tab:blue', 'tab:orange'], 
           height = 7)

# Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação Entre Altura e Peso de Fumantes e Não Fumantes')

# Remove as bordas
sns.despine()

# Show
plt.show()
```
